**Importing the libraries**

```python
import numpy as np
import pandas as pd

from sklearn.utils import resample
```

**loading the dataset from the csv file with the help of pandas**

```python
df=pd.read_csv("spam.csv",encoding="latin1")
df
```

```python

# df.info()
df=df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python

df.isnull().sum()

```

```python
x=df.v2
```

```python
y=df.v1
```

```python
y.replace(to_replace='ham',value=1,inplace=True)
y.replace(to_replace='spam',value=0,inplace=True)
```

```python
y=df.v1
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=19)
```

```python
gok= pd.concat([x_train, y_train], axis=1)

# separate minority and majority classes
spam = gok[gok.v1==0]
ham = gok[gok.v1==1]
```

```python
assem=resample(spam,replace=True,n_samples=len(ham),random_state=25)
```

```python
upsample=pd.concat([ham,assem])
upsample.v1.value_counts()
```

```python
y_train=upsample.v1

x_train=upsample.v2
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
tf=vect.fit_transform(x_train)
tt=vect.transform(x_test)

y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(tf)
```

```python
mail=[
    'hey i am the hacker'
]

mail=vect.transform(mail)
print(mail)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()

model.fit(tf,y_train)
```

```python
model.predict(tt)
```

```python
model.score(tt,y_test)
```

```python
model_predictions = model.predict(mail)
```

```python
import gzip, pickle
with gzip.open('model.pklz', 'wb') as ofp:
    pickle.dump(model, ofp)
```

```python
mail_id=['Hurry in 23  a daily comp to win FA Cup final tickets']
input_data_features=vect.transform(mail_id)
```

```python
prediction=mod.predict(input_data_features)
```
